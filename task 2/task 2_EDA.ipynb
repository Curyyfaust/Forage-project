```python
#client--churn
print(df_client["churn"].value_counts())
plt.figure(figsize=(12,4))
sns.countplot(data=df_client,x="churn")
plt.show()
```

```python
#Sales channel
channel=df_client[["churn","channel_sales"]]
plt.figure(figsize=(12,8))
for i,column in enumerate(channel["channel_sales"].unique(),1):
    plt.subplot(2,4,i)
    sns.countplot(data=channel[channel["channel_sales"]==column],x="churn")
    plt.title(column)
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Consumption
consumption=df_client[["cons_12m","cons_gas_12m","cons_last_month","imp_cons","churn"]]
columns=["cons_12m","cons_gas_12m","cons_last_month","imp_cons"]
plt.figure(figsize=(12,6))
for i,column in enumerate(columns,1):
    #print(i,column) #10
    plt.subplot(3,2,i)
    #sns.displot(x="y",y=column,data=df,palette="rainbow")
    sns.kdeplot(x=consumption[(consumption["churn"]==1)][column])
    sns.kdeplot(x=consumption[(consumption["churn"]==0)][column])
    plt.title(column,fontsize=14,fontweight="bold")
    plt.xlabel("Value")
    plt.ylabel("Freq")
    plt.legend(["1","0"])

plt.tight_layout()
```

```python
#Forecast
forecast = df_client[
    ["forecast_cons_12m",
    "forecast_cons_year","forecast_discount_energy","forecast_meter_rent_12m",
    "forecast_price_energy_off_peak","forecast_price_energy_peak",
    "forecast_price_pow_off_peak","churn"
    ]
]
columns =  ["forecast_cons_12m",
    "forecast_cons_year","forecast_discount_energy","forecast_meter_rent_12m",
    "forecast_price_energy_off_peak","forecast_price_energy_peak",
    "forecast_price_pow_off_peak"
    ]
plt.figure(figsize=(12,8))
for i,column in enumerate(columns,1):
    #print(i,column) #10
    plt.subplot(4,2,i)
    #sns.displot(x="y",y=column,data=df,palette="rainbow")
    sns.kdeplot(x=forecast[(forecast["churn"]==1)][column])
    sns.kdeplot(x=forecast[(forecast["churn"]==0)][column])
    plt.title(column)
    plt.xlabel("Value")
    plt.ylabel("Freq")
    plt.legend(["1","0"])

plt.tight_layout()
```
